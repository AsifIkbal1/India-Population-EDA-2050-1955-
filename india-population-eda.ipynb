```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from matplotlib import pyplot as plt
import warnings
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
warnings.filterwarnings("ignore")
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
india = pd.read_csv("/kaggle/input/population-data-india/india_population.csv")
india.info()
```

```python
india.describe(include='all')
```

# EDA

```python
_,ax = plt.subplots(len(india.columns[1:]),1,figsize=(15,35))
for i,val in enumerate(india.columns[1:]):
    ax[i].bar(india["Year"],india[val])
    ax[i].set_ylabel(val)
plt.show()
```

# Preprocessing

```python
scaler_features = MinMaxScaler().fit_transform(india[india.columns.values[1:]])
for i,val in enumerate(india.columns.values[1:]):
    india[val+"_scaled"] = scaler_features[:,i]
india.describe(include='all')
```

```python
ax = india[['Population_scaled','Migrants (net)_scaled','Fertility Rate_scaled','Median Age_scaled','Urban Population_scaled','World Population_scaled']].plot(kind='bar',figsize=(20,5))
ax.set_xticklabels(india['Year'].values)
```

```python
#population change next to median age and fertility rate
ax = india[['Population_scaled','Fertility Rate_scaled','Median Age_scaled']].plot(kind='bar',stacked=True,figsize=(20,5))
ax.set_xticklabels(india['Year'].values)
```

```python
#population change next to migrants and urban populations
ax = india[['Population_scaled','Migrants (net)_scaled','Urban Population_scaled']].plot(kind='bar',stacked=True,figsize=(20,5))
ax.set_xticklabels(india['Year'].values)
```

```python
#india and world population
ax = india[['Population','World Population']].plot(kind='bar',figsize=(15,5))
ax.set_xticklabels(india['Year'].values)
```

# Statistics

```python
sns.distplot(india['Population_scaled'])
plt.show()
```

```python
india.boxplot(column=india.columns.values[13:].tolist(),figsize=(15,5),widths=0.3)
plt.xticks(rotation = 45) 
plt.show()
```

```python
_,ax = plt.subplots(5,1,figsize=(10,10))
sns.regplot(x='Population_scaled',y='Median Age_scaled',data=india,order=2,ax=ax[0])
sns.regplot(x='Population_scaled',y='Fertility Rate_scaled',data=india,order=2,ax=ax[1])
sns.regplot(x='Population_scaled',y='Urban Population_scaled',data=india,order=2,ax=ax[2])
sns.regplot(x='Population_scaled',y='Migrants (net)_scaled',data=india,order=2,ax=ax[3])
sns.regplot(x='Population_scaled',y='World Population_scaled',data=india,order=2,ax=ax[4])
plt.show()
```
